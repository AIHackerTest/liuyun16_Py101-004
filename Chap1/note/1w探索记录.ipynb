### 1. Dict的学习
 - 定义：`Dict={key1:value1, key2:value2}`，keys的数据类型不能为list，dict或其他mutable数据类型，必须hashable。
 
 - 一些方法：
    - d[key]。检索出对应的value
    - d[key]=value，赋值；
    - del d[key]
    - key in/not in d
    - iter(d)?:Return an iterator over the keys of the dictionary. This is a shortcut for iter(d.keys()).
    - copy(): Return a **shallow copy** of the dictionary.
    - get(key):Return the value for key if key is in the dictionary, else default.
    - items():Return a new view of the dictionary’s items ((key, value) pairs).

##### 学习字符编码
1. 英文字母和数字以及一般的符号，用ASCII码就可以进行字符-二进制码的索引；
字母A用ASCII编码是十进制的`65`，二进制的`01000001`；

字符0用ASCII编码是十进制的`48`，二进制的`00110000`，注意字符`'0'`和整数`0`是不同的；
2. 为了调解各国语言之间可识别（不出现乱码），如中文，韩文，日文，囊括到计算机可识别的编码，出现了Unicode编码。
汉字中已经超出了ASCII编码的范围，用Unicode编码是十进制的20013，二进制的01001110 00101101。
但是，**ASCII编码是1个字节，而Unicode编码通常是2个字节。**如果统一成Unicode编码，乱码问题从此消失了。但是，用Unicode编码比ASCII编码需要多一倍的存储空间，因此对于纯英文文本，在存储和传输上就十分不划算。
为了节约存储空间，又出现了把Unicode编码转化为**“可变长编码”的UTF-8编码**。UTF-8编码把一个Unicode字符根据不同的数字大小编码成1-6个字节，**常用的英文字母被编码成1个字节，汉字通常是3个字节**，只有很生僻的字符才会被编码成4-6个字节。如果你要传输的文本包含大量英文字符，用UTF-8编码就能节省空间：

字符	ASCII	Unicode	UTF-8
A	01000001	00000000 01000001	01000001
中	x	01001110 00101101	11100100 10111000 10101101

```python
"""
将一行为“北京，晴”格式的文本转换为字典{"北京":晴}
"""
city_dic={}
f_w=open('C:/git/Py101-004/Chap1/resource/weather_info.txt', 'r',encoding='utf-8')
p=f_w.readline().strip('\n') #得到的字符串的内容去掉\n

p1=p.split(",") #将得到的字符串内容，用逗号，隔开为多个字符串的list

city_dic[p1[0]]=p1[1]# 构建city-weather作为key-value的dict。

print(p,p1,city_dic)

f_w.close()
```
